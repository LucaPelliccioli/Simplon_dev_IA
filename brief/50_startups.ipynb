```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import sklearn.linear_model
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error

startups_stats = pd.read_csv("/home/simplon/Simplon_dev_IA/jeux_de_données/ML/50_Startups.csv")
startups_stats
```

```python
# Prepare the data
X = np.c_[startups_stats[["R&D Spend","Administration","Marketing Spend"]]]
y = np.c_[startups_stats["Profit"]]

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.4, random_state=0)

X_valid, X_test, y_valid, y_test = train_test_split(
    X_test, y_test, test_size=0.5)

print(X.shape)

print(X_train.shape)
print(X_valid.shape)
print(X_test.shape)

# Visualize the data
startups_stats.plot(kind='scatter', x="R&D Spend", y='Profit')
plt.show()

# Select a linear model
model = sklearn.linear_model.LinearRegression()

# Train the model
model.fit(X_train, y_train)
```

```python
y_true = y_test
y_pred = model.predict(X_test)
print(model.predict(X_test))
print(mean_absolute_error(y_true, y_pred))
```
