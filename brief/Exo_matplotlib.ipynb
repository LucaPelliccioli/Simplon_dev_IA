# Brief Matplotlib Partie 2

```python
import pandas as pd
import matplotlib as mtp
import matplotlib.pyplot as plt

unrate = pd.read_csv("../jeux_de_données/unrate.csv", parse_dates= ["DATE"])
unrate
```

### Introduction

```python
unrate["DATE"] = pd.to_datetime(unrate["DATE"])
unrate["DATE"]
```

### 1. Premier graphique

```python
plt.figure(figsize=[15,6], dpi=75)

plt.plot(unrate["DATE"][0:12], unrate["UNRATE"][0:12])
plt.show()
plt.close()
```

### 2. Créer une rotation sur les données de l'axe x

```python
plt.figure(figsize=[15,7], dpi=75)

plt.plot(unrate["DATE"][0:12], unrate["UNRATE"][0:12], "b.-")

plt.xticks(rotation="vertical")
plt.show()
plt.close()
```

### 3. Ajouter les intitulés des axes et un titre à notre graphique

```python
plt.figure(figsize=[15,6], dpi=75)

plt.plot(unrate["DATE"][0:12], unrate["UNRATE"][0:12], "b.-")

plt.title("Evolution du taux de chômage en 1948 aux USA", fontdict={"fontweight" : "bold", "fontsize" : 20})
plt.ylabel("Taux de chômage (%)")
plt.xlabel("Mois")

plt.show()
plt.close()
```

### 4.& 5. Position des graphiques

```python
fig, axs = plt.subplots(2,3, figsize=(30,15), dpi=90)

valeur_date = unrate["DATE"][:12]
valeur_unrate = unrate["UNRATE"][:12]
valeur_date2 = unrate["DATE"][12:24]
valeur_unrate2 = unrate["UNRATE"][12:24]

axs[0,0].plot(valeur_date, valeur_unrate, "b.-")
axs[0,0].set_title("Evolution du taux de chômage en 1948 aux USA")
axs[0,0].set_ylabel("Taux de chômage en 1948")
axs[0,0].set_xlabel("Mois")

axs[1,0].plot(valeur_date2, valeur_unrate2, "r.-")
axs[1,0].set_title("Evolution du taux de chômage en 1949 aux USA")
axs[1,0].set_ylabel("Taux de chômage en 1949")
axs[1,0].set_xlabel("Mois")

plt.xticks(rotation=90)
plt.show()
plt.close()
```

### 6. Superposition de courbes

```python
unrate['MONTH'] = unrate['DATE'].dt.month
```

```python
plt.figure(figsize=[16,5], dpi=75)
plt.subplot(111)

plt.plot(unrate[0:12]["MONTH"], unrate[0:12]["UNRATE"], "r.-", label="Taux 1948")
plt.plot(unrate[12:24]["MONTH"], unrate[12:24]["UNRATE"], "b.-", label="Taux 1949")

plt.title("Evolution du taux de chômage aux USA")

plt.ylabel("Taux de chômage (%)")
plt.xlabel("Mois")
plt.show()
plt.close()
```

### 7. Ajouter plus de courbes

```python
step = 12
mycolor = ['r','b','g','C1','k']
unrate['DATE'] = pd.to_datetime(unrate['DATE'])
unrate['MONTH'] = unrate['DATE'].dt.month
plt.figure(figsize=(10, 6))
plt.xticks(rotation=90)

for n in range(0,5):

    myannee = 1948 + n
    plt.plot(unrate["MONTH"][0+n*step:(1+n)*step],unrate["UNRATE"][0+n*step:(1+n)*step],mycolor[n],label=myannee,markersize=1.8)


plt.title('Evolution du taux de chômage de 1948 à 1952')
plt.ylabel('Taux de chômage (en %)')
plt.xlabel('Mois (de 1 à 12)')
plt.legend(loc='upper left')

# à ne pas confondre avec annee
#        [0+nstep:(1+n)step] avec n allant de 0 à 5 et step = 12
# n = 0 , premier tour -> [0+0step:(1+0)step] -> [0:step] -> [0:12]
# n = 1 , deuxieme tour -> [0+1step:(1+1)step] -> [1step:2step] -> [12:24]
# n = 2 , troisieme tour -> [0+2step:(2+1)step] -> [2step:3step] -> [24:36]
# etc....
```
