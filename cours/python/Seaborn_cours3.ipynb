```python
import pandas as pd
import numpy as np
import matplotlib as mtp
import matplotlib.pyplot as plt
import seaborn as sns

donnees = pd.read_csv("../../jeux_de_données/120-years-of-olympic-history-athletes-and-results.csv", index_col=[1])
donnees

# Création du nouveau Dataset sans les lignes contenant les valeurs manquantes
donnees_NaN_supprimes = donnees.dropna()
donnees_NaN_supprimes = donnees_NaN_supprimes.drop_duplicates()
donnees_NaN_supprimes.isnull().sum()

donnees_NaN_remplaces = donnees.drop_duplicates()
donnees_NaN_remplaces

donnees['Medal']=donnees['Medal'].fillna("None")
donnees_NaN_supprimes = donnees.dropna()
donnees_NaN_supprimes = donnees_NaN_supprimes.drop_duplicates()
donnees_NaN_supprimes.isnull().sum()
moyennes=donnees[['Age','Height','Weight']].mean()
donnees[['Age','Height','Weight']]=donnees[['Age','Height','Weight']].fillna(moyennes)
donnees_NaN_remplaces=donnees.drop_duplicates()

print("Tableau avec NaN remplacés")
print(donnees_NaN_remplaces.shape)
print("Tableau avec NaN supprimés")
print(donnees_NaN_supprimes.shape)
```

## 3.8. Histogrammes

```python
plt.figure()
plt.figure(figsize = (8,5), dpi=70)

palette = sns.color_palette(["#3498db", "#e74c3c"])
sns.set_palette(palette)

plt.subplot()
sns.distplot(donnees_NaN_supprimes["Height"],kde=True, bins=20)
plt.title("Distribution des tailles chez les athlètes des JO")
plt.xticks(rotation=90)
plt.show()
plt.close()
```

```python
plt.figure()
plt.figure(figsize = (8,5), dpi=70)
plt.subplot()

palette = sns.color_palette(["#3498db", "#e74c3c"])
sns.set_palette(palette)

sns.histplot(donnees_NaN_supprimes["Height"],kde=False, bins=20)
plt.title("Distribution des tailles chez les athlètes des JO")
plt.xticks(rotation=90)
plt.show()
plt.close()
```

```python
plt.figure()
plt.figure(figsize = (8,5), dpi=70)

valeurs_athletes_feminines = donnees_NaN_supprimes.loc[(donnees_NaN_supprimes["Sex"]=="F")]
valeurs_athletes_masculines = donnees_NaN_supprimes.loc[(donnees_NaN_supprimes["Sex"]=="M")]

plt.subplot()
sns.distplot(valeurs_athletes_feminines["Height"],label="Tailles F")
sns.distplot(valeurs_athletes_masculines["Height"],label="Tailles M")

plt.title("Distribution des tailles chez les athlètes des JO")
plt.xticks(rotation=90)
plt.legend()
plt.show()
plt.close()
```

## 3.9. Joinplot

```python
plt.figure()
plt.figure(figsize = (8,5), dpi=70)

sns.set(style="white",color_codes=True)
sns.jointplot(data=donnees_NaN_supprimes,x="Height",y="Weight")
plt.show()
plt.close()
```

```python
plt.figure()
plt.figure(figsize = (8,5), dpi=70)

sns.set(style="white",color_codes=True)
sns.jointplot(data=donnees_NaN_supprimes,x="Height",y="Weight", kind="reg",line_kws={"color":"red"})
plt.show()
plt.close()
```
