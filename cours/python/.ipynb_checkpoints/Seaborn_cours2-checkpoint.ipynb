```python
import pandas as pd
import numpy as np
import matplotlib as mtp
import matplotlib.pyplot as plt
import seaborn as sns

donnees = pd.read_csv("../../jeux_de_données/120-years-of-olympic-history-athletes-and-results.csv", index_col=[1])
donnees

# Création du nouveau Dataset sans les lignes contenant les valeurs manquantes
donnees_NaN_supprimes = donnees.dropna()
donnees_NaN_supprimes = donnees_NaN_supprimes.drop_duplicates()
donnees_NaN_supprimes.isnull().sum()

donnees_NaN_remplaces = donnees.drop_duplicates()
donnees_NaN_remplaces

donnees['Medal']=donnees['Medal'].fillna("None")
donnees_NaN_supprimes = donnees.dropna()
donnees_NaN_supprimes = donnees_NaN_supprimes.drop_duplicates()
donnees_NaN_supprimes.isnull().sum()
moyennes=donnees[['Age','Height','Weight']].mean()
donnees[['Age','Height','Weight']]=donnees[['Age','Height','Weight']].fillna(moyennes)
donnees_NaN_remplaces=donnees.drop_duplicates()

print("Tableau avec NaN remplacés")
print(donnees_NaN_remplaces.shape)
print("Tableau avec NaN supprimés")
print(donnees_NaN_supprimes.shape)
```

## 2.4. Point Plot

```python
plt.figure()
plt.figure(figsize = (10,8), dpi=70)
plt.subplot()

sns.set_context("notebook", rc={"xtick.labelsize" : 15, "ytick.labelsize" : 15})
sns.set_style("white")

donnees_summer = donnees_NaN_remplaces[donnees_NaN_remplaces["Season"] == "Summer"]
donnees_pointplot = donnees_summer.groupby(["Year","Sex"]).size().reset_index(name="Count")

sns.pointplot(x="Year", y="Count", data = donnees_pointplot, hue="Sex")
plt.xticks(rotation=90)
plt.show()
plt.close()
```

```python
plt.figure()
plt.figure(figsize = (10,5), dpi=70)
plt.subplot()

sns.set_context("notebook",rc={"xtick.labelsize" : 15, "ytick.labelsize" : 15})

donnees_summer = donnees_NaN_remplaces[donnees_NaN_remplaces["Season"] == "Summer"]

sns.pointplot(x="Year", y="Age", data=donnees_summer,hue="Sex")
plt.xticks(rotation=90)
plt.show()
plt.close()
```

```python
plt.figure()
plt.figure(figsize = (10,5), dpi=70)
plt.subplot()

donnees_summer = donnees_NaN_supprimes[donnees_NaN_supprimes["Season"]=="Summer"]

palette = sns.color_palette(["#3498db", "#e74c3c"])
sns.set_palette(palette)

sns.stripplot(x="Year", y="Height",hue="Sex", data=donnees_summer)
#sns.stripplot(x="Medal",y="Height",hue="Sex",data=donnees_summer)
plt.xticks(rotation=90)
plt.show()
plt.close()
```

## 3.6. Boxplot

On constate que les athlètes avec l’âge médian le plus élevé appartiennent aux disciplines bobsleigh, curling et skeleton (un sport de glisse sur une planche ressemblant à une luge). Ce sont aussi les trois sports pour lesquels la distribution de l’âge est la plus large autour de la médiane. Le sport pour lequel l’âge médian est le plus bas est la discipline Figure Skating, qui signifie Patinage artistique en français. Il est connu que dans cette discipline, les médaillés d’or sont très jeunes, contrairement à ce qu’on peut voir en Curling par exemple, où il n’est pas rare que le médaillé d’or ait dépassé la trentaine.

```python
plt.figure()
plt.figure(figsize = (10,5), dpi=70)

donnees_winter = donnees_NaN_supprimes[donnees_NaN_supprimes["Season"]=="Winter"]

plt.figure(figsize=[15,6])

sns.boxplot(x="Sport",y="Age",data=donnees_winter)
plt.xticks(rotation=90)
plt.show()
plt.close()
```

## 3.7. Graphique à barres : countplot

```python
plt.figure()
plt.figure(figsize = (10,5), dpi=70)
plt.subplot()

gold = (donnees_NaN_supprimes["Medal"]=="Gold")
donnees_medaille_or = donnees_NaN_supprimes.loc[gold]

sns.countplot(data=donnees_medaille_or,x="Age")
plt.title("Distribution de l'age des athlètes ayant obtenu une médaill d'or")
plt.xticks(rotation=90)
plt.show()
plt.close()
```

```python
plt.figure()
plt.figure(figsize = (10,5), dpi=70)

palette = sns.color_palette("Paired")
sns.set_palette(palette)

gold45 = (donnees_NaN_supprimes["Medal"]=="Gold") & (donnees_NaN_supprimes["Age"] > 45)
gold15 = (donnees_NaN_supprimes["Medal"]=="Gold") & (donnees_NaN_supprimes["Age"] < 15)
disciplines_ages = donnees_NaN_supprimes.loc[gold45]
disciplines_jeunes = donnees_NaN_supprimes.loc[gold15]

plt.subplot(121)
sns.countplot(data=disciplines_ages, x="Sport")
plt.title("Disciplines des médailles d'or de plus de 45 ans")
plt.xticks(rotation=90)

plt.subplot(122)
sns.countplot(data=disciplines_jeunes,x="Sport")
plt.title("Disciplines des médaillés d'or de moins de 15 ans")
plt.xticks(rotation=90)

plt.show()
plt.close()
```
